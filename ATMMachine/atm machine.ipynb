```python
import win32com.client as wincom
speak = wincom.Dispatch("SAPI.SpVoice")


class atm():
    def __init__(self):
        self.pin= ""
        self.balance = 0
        self.menu()

    def menu(self):
        speak.speak("hello, select Desired option")
        
        userInput = input(
       """ hello, enter our operation
        *Enter 1 for create pin
        *enter 2 for deposit
        *Enter 3 for withdraw
        *Enter 4 for Check Balance
        *Enter 5 for exit""")

        
        if userInput == "1":
            self.createpin()
        elif userInput == "2":
            self.deposit()
        elif userInput == "3":
            self.withdraw()
        elif userInput == "4":
            self.checkbal()
        elif userInput == "5":
            print("bye")
        else:
            print("Invalid input. Please try again.")
            self.menu()
                

    # create pin
    def createpin(self):
        try:
            self.pin = input("create pin")
        except Exception as e:
            print("An error occurred:", e)
        else:
            self.menu()

    # Deposit
    def deposit(self):
        try:
            speak.speak("Enter your pin")
            pi = input("enter your pin: ")
            if pi == self.pin:
                speak.speak("Enter Amount for deposit")
                damt = int(input("Enter Amount for deposit: "))
                self.balance += damt
            else:
                print("invalid pin")
                print("Please Redo")
        except Exception as e:
            print("An error occurred:", e)
        finally:
            self.menu()

    # Withdraw
    def withdraw(self):
        try:
            speak.speak("Enter your pin")
            pi = input("Enter your pin: ")
            if pi == self.pin:
                speak.speak("Enter Amount for withdraw")
                am = int(input("Enter amount for withdraw: "))
                if am <= self.balance:
                    self.balance -= am
                else:
                    print("insufficient balance")
                    print("Please Redo")
            else:
                print("invalid pin")
                print("Please Redo")
        except Exception as e:
            print("An error occurred:", e)
        finally:
            self.menu()

    # Check Balance
    def checkbal(self):
        speak.speak("Enter your pin")
        pi = input("Enter your pin")
        if pi == self.pin:
            speak.speak(f"Your balance is {self.balance}")
            print("Your balance is: ", self.balance)
        else:
            print("invalid pin")
            print("Please Redo")
            self.menu()

sbi = atm()
```

```python
import tkinter as tk
from tkinter import messagebox
import win32com.client as wincom

speak = wincom.Dispatch("SAPI.SpVoice")

class BikeShop:
    def __init__(self, stock):
        self.stock = stock

    def displayBike(self):
        speak.Speak(f"Total Bikes in Stock are: {self.stock}")
        print("Bikes in Stock are: ", self.stock)

    def bikeForRent(self, q):
        if q <= 0:
            speak.Speak("Enter a positive value greater than zero")
            print("Enter a positive value greater than zero")
        elif q > self.stock:
            speak.Speak("Entered value is greater than current available stock")
            print("Entered value is greater than current available stock")
        else:
            total = q * 100
            speak.Speak(f"Total Rental Price is: {total}")
            print("Total Rental Price is: ", total)
            result = messagebox.askyesno("Confirmation", f"Do you want to rent {q} bikes?")
            if result:
                self.stock = self.stock - q
                speak.Speak(f"Thank You, You can pick {q} bikes from the store")
                print(f"You can pick {q} bikes from the store")

obj = BikeShop(100)

def display_stock():
    obj.displayBike()

def rent_bike():
    n = int(entry.get())
    obj.bikeForRent(n)

# Create GUI
root = tk.Tk()
root.title("Manusmare Bike Shop")

label = tk.Label(root, text="Welcome To Manusmare Bikes")
label.pack()

label2 = tk.Label(root, text="Please select the desired option")
label2.pack()

button1 = tk.Button(root, text="Display Stock", command=display_stock)
button1.pack()

label3 = tk.Label(root, text="Enter the quantity of required bikes:")
label3.pack()

entry = tk.Entry(root)
entry.pack()

button2 = tk.Button(root, text="Rent a Bike", command=rent_bike)
button2.pack()

button3 = tk.Button(root, text="Exit", command=root.quit)
button3.pack()

root.mainloop()
```
